##  1.Keep top 3 frequent values as it is in the series and replace others values by average[Series should have minimum 3 distinct elements].


```python
series = pd.Series([1,3,5,5,6,6,7,7,7])
freq = series.value_counts()
series[~series.isin(freq .index[:3])] = series.mean()
print(series)
```

## 3.a Convert a numpy array to a dataframe of given shape [User given]


```python
print("Enter the dimensions for the dataframe: ")
row=int(input("Row: "))
col=int(input("Column: "))

lis=[[0 for i in range(col)] for j in range(row)]

print("Enter", row*col, "Elements for the dataframe")

for i in range(row):
    for j in range(col):
        lis[i][j]=int(input())

# Numpy Array        
nparr=np.array(lis)

# dataframe
df = pd.DataFrame(nparr, columns = [("Column_"+ f'{i+1}' ) for i in range(col)])
df
```

## 3.b Calculate the number of characters in each word in a series [ User given, series should have at least 5 elements]


```python
n = int(input("Enter the number of the names you want to input: "))
name = []
for i in range(0, n):
    ele = input("Enter name: \n")
    name.append(ele)
words = pd.Series(name)
rst = words.map(lambda calc: len(calc))
print("No. of characters in each word in the given series:")
print(rst)
```

## 4.4. Use this dataframe and find out the mean and highest mark for each class.



```python
data = {'Class':['A', 'A', 'B', 'B', 'A', 'A', 'B', 'B'],
        'Mark':[1, 2, 3, 4, 5, 6, 7, 8]}
df = pd.DataFrame(data)
df
```

```python
#mean of each class
mean_data = df.groupby("Class")['Mark'].agg({"mean"})
mean_data.head()
```

```python
#Highest mark in each class
grouped_df = df.groupby("Class")
highest_mark = grouped_df.max()
highest_mark = highest_mark.reset_index()
print(highest_mark)
```

## 4.Find the desired output as shown in the image below.


```python
lis=[["Shingeki No Kyojin",["Action","Drama","Fantasy","Military","Mystery","Shounen","Super Power"]],
     ["Full Metal Alchemist: Brotherhood",["Action","Military","Adventure","Comedy","Drama","Magic","Fantasy","Shounen"]]]

df=pd.DataFrame(lis, columns=["Anime", "Genre"])
```

### Output 1: (Exception: for loop can be used for this question)

```python
def Union(lst1, lst2):
    final_list = list(set(lst1) | set(lst2))
    return final_list
genres=Union(df.at[0,"Genre"],df.at[1,"Genre"])

for i in genres:
    if i in df.at[0,"Genre"]:
        df.loc[0,i]=1
    else:
        df.loc[0,i]=0
    if i in df.at[1,"Genre"]:
        df.loc[1,i]=1
    else:
        df.loc[1,i]=0
        
df
```
