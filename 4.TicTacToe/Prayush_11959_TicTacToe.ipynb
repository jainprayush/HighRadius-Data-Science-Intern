```python
def get_input(message):
  while True:
    try:
      num = int(input(message))
      if num >= 1 and num <= 9:
        return num
      elif num == 0:
        return 0;
      else:
        print('>> Input not valid \n ')
    except:
      print('>> Input not valid \n ')
```

```python
def get_auto_input(message):
  while True:
    try:
      choice = input(message)
      if choice == 'mark' or choice == 'MARK':
        return 'continue'
      elif choice == 'quit' or choice == 'q' or choice == 'QUIT' or choice == 'Q':
        return 'end'
      else:
        print('>> Input not valid \n ')
    except:
      print('>> Input not valid \n ')
```
